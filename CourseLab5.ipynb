# Example:
### Write a python program that takes two interger numbers from the user. Then it divides the first number by the second number if the second number is not equals to zero, otherwise, it prints "cannot divide by zero".

```python
num1=int(input("enter the first number: "))
num2 = int(input("enter the second number: "))
if num2==0: 
    print("cannot divide by zero")
else: 
    print(num1/num2)
print('finish')
```

### The previous if/else statement can be rewritten as:
### <font color=blue>if condition:Block1 </font>
### <font color=blue>else: Block2</font>

```python
num1=int(input("enter the first number: "))
num2 = int(input("enter the second number: "))
if num2==0: print("cannot divide by zero")
else: print(num1/num2)
print('finish')
```

### If  the previous if/else statement contain one statement in Block1 and one statement in Block2,  then it can be rewritten as:
## <font color=blue> Block1 if condition else Block2 </font>
### It means:  perform Block1 if the condition is true otherwise perform Block2
### This is known as <font color=red> Inline If </font>

```python
num1=int(input("enter the first number: "))
num2 = int(input("enter the second number: "))
print("cannot divide by zero") if num2==0 else print(num1/num2)
print('finish')
```

# <font color=red> Multi-way Decision Statements(if/elif/else statement)</font>
### Is to take one of many decisions.

```python
value = int(input("Plz. enter an integer in the range 0...5: "))
if value < 0:
    print("Too small")
else:
    if value == 0:
        print("zero")
    else:
        if value == 1:
            print("one")
        else:
            if value == 2:
                print("two")
            else:
                if value == 3:
                    print("three")
                else:
                    if value == 4:
                        print("four")
                    else:
                        if value == 5:
                            print("five")
                        else:
                            print("Too large")
print("Done")
```

### The previous program can be rewitten using eiif... where every else if is replace with elif

```python
value = int(input("Plz. enter an integer in the range 0...5: "))
if value < 0:
    print("Too small")
elif value == 0:
    print("zero")
elif value == 1:
    print("one")
elif value == 2:
    print("two")
elif value == 3:
    print("three")
elif value == 4:
    print("four")
elif value == 5:
    print("five")
else:
    print("Too large")
print("Done")
```

## <font color=red> While Loop</font>


### <font color=blue>Example:</font> Write a python program that prints numbers from 1 to 5

```python
count = 1 # Initialize counter
while count <= 5: # Should we continue?
    print(count) # Display counter, then
    count += 1 # Increment counter
print('finish')
```

### <font color=blue>Example:</font> Write a python program that ask the user to input two integer numbers (a, b) then it calculate and prints ($a^b$) , i.e., result=a\*a\*a\*a.....(b times)

```python
a=int(input("please enter an integer value: "))
b=int(input("please enter an integer value: "))
i=1
exponent=1
while i<=b:
    exponent*=a
    i+=1
print(exponent)
```

### <font color=blue>Example:</font> Write a python program that ask the user to input two integer numbers (a, b) then it calculate and prints the sum and the average of the numbers from a to b

```python
a=int(input("please enter an integer value: "))
b=int(input("please enter an integer value: "))
i=a
sum=0
while i<=b:
    sum+=i
    i+=1
print("summation is {}, and the average is{}".format(sum, sum/(b-a+1)))
```

# <font color=red>For Loop</font>
### A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).
### With the for loop we can execute a set of statements, one for each item in a list, tuple, set etc.

## <font color=blue>To print a list value</font>
#### l=[5, 8, 2, 7, 3]
#### print(l[0])
#### print(l[1])
#### print(l[2])
#### print(l[3])
#### print(l[4])

#### Thus, print(l[i]) where i from 0 to 4

### <font color=blue>To print a list value using while loop</font> 

```python
l=[5, 8, 2, 7, 3]
i=0
while i<=4:
    print(l[i])
    i=i+1
print('finish')
```

### <font color=blue>To print a list value using for loop</font>   

```python
l=[5, 8, 2, 7, 3]
for i in l:     # i will iterate over each item in the list l
    print(i)
print('finish')
```

## <font color=blue>break</font>  is used to exit from the loop

```python
j = 0
for i in range(5):
    j = j + 2
    print ('i = ', i, ', j = ', j)
    if j == 6:
        break
print('Done')
```

## <font color=blue>continue</font>  skip this iteration and continue the rest of the loop

```python
for x in range(10):   # Check if x is even
    if x % 2 == 0:
        continue
    print(x)
print('Done')
```
